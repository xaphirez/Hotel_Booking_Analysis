## Data Cleaning

### Missing Value Analysis
After inspecting the dataset, the following columns contain missing values:

| Column | Missing Values | Percentage | Action |
|---|---|---|---|
| children | 4 | 0.00% | Fill with 0 |
| country | 488 | 0.41% | Fill with "Unknown" |
| agent | 16,340 | 13.69% | Fill with 0 |
| company | 112,593 | 94.31% | Fill with 0 |

### Decisions

**Children**
Only 4 rows are missing — less than 0.01% of the dataset. 
Since no children is the most logical default for a hotel booking, 
missing values are filled with 0 rather than dropping the rows.

**Country**
488 rows (0.41%) have no country recorded. Rather than dropping 
these rows, missing values are filled with "Unknown" to acknowledge 
that the guest's origin was not captured — possibly due to privacy 
preferences or data entry omission. The low percentage means this 
will not significantly impact analysis.

**Agent**
13.69% of bookings have no agent ID recorded. This most likely 
indicates that the booking was made directly without going through 
a travel agent. Missing values are filled with 0 to represent 
"no agent" and distinguish these bookings from agent-assisted ones.

**Company**
94.31% of bookings have no company ID — by far the largest missing 
value in the dataset. This is expected behavior rather than a data 
quality issue, as the majority of hotel guests are individual 
travelers booking for personal reasons rather than on behalf of 
a company. Missing values are filled with 0 to represent 
"no company affiliation."

```python
#Fill children missing values with 0
df['children'] = df['children'].fillna(0)
#Fill Country missing values with unknown
df['country'] = df['country'].fillna('Unknown')
#Fill agent and company missing values with 0 (it means self booked)
df['agent']=df['agent'].fillna(0)
df['company']=df['company'].fillna(0)
```

### Duplicate Data

The dataset contains a high number of duplicate rows, accounting for 
26.8% of the total data. Upon further investigation, there is no unique 
identifier column (such as a booking ID) to differentiate records, 
which makes it difficult to determine whether duplicates represent 
legitimate entries.

After inspecting the duplicate rows, some records appear exactly twice 
and others appear three times across all 32 columns. In a real hotel 
booking context, it is highly unlikely for the same guest to make 
identical bookings multiple times on the same date at the same rate. 
This strongly suggests the duplicates are the result of a data export 
or merging error rather than actual bookings.

Therefore, duplicate rows have been removed to prevent inflating 
metrics and ensure the integrity of the analysis.

```python
df = df.drop_duplicates()

print(f"data frame rows after removing duplicates: {df.shape[0]}")
```

```python
# Bookings with zero guests
zero_guests = df[(df['adults'] == 0) & (df['children'] == 0) & (df['babies'] == 0)]
print(f"Bookings with zero guests: {len(zero_guests)}")

# Bookings with zero nights
zero_nights = df[(df['stays_in_weekend_nights'] == 0) & (df['stays_in_week_nights'] == 0)]
print(f"Bookings with zero nights: {len(zero_nights)}")

# Bookings with negative or zero ADR
invalid_adr = df[df['adr'] <= 0]
print(f"Bookings with invalid ADR: {len(invalid_adr)}")
```

```python
# Separate negative vs zero ADR
print(f"Zero ADR: {len(df[df['adr'] == 0])}")
print(f"Negative ADR: {len(df[df['adr'] < 0])}")
```

### Invalid Data Removal

A small number of bookings were found to contain invalid values 
across three columns:

- **Zero guests (166 rows):** Some bookings recorded zero adults, 
  zero children, and zero babies simultaneously. A hotel booking 
  with no guests is logically impossible and is most likely the 
  result of a data entry error.

- **Zero nights (651 rows):** Some bookings recorded zero weekend 
  nights and zero weekday nights simultaneously, meaning the guest 
  had no actual stay duration. These are likely same-day cancellations 
  or data entry errors.

- **Invalid ADR (1,779 rows):** Some bookings recorded an Average 
  Daily Rate of zero or below. While zero ADR could theoretically 
  represent complimentary stays, we cannot confirm this without 
  additional context. Negative ADR is impossible in any scenario. 
  To maintain the integrity of revenue analysis, all rows with 
  ADR equal to or below zero have been removed.

All three categories of invalid data have been removed from the 
dataset to ensure accurate and reliable analysis.

```python
# Step 1 — at least 1 adult
df_clean = df[df['adults'] > 0]

# Step 2 — at least 1 night
df_clean = df_clean[(df_clean['stays_in_weekend_nights'] > 0) | (df_clean['stays_in_week_nights'] > 0)]

# Step 3 — valid ADR
df_clean = df_clean[df_clean['adr'] > 0]

# Check result
print(f"Rows after removing invalid data: {df_clean.shape[0]}")
```

## Question 4 — Room Upgrade Percentage

```python
# Count True vs False
df_clean['is_room_upgraded'].value_counts()

# Percentage
df_clean['is_room_upgraded'].value_counts(normalize=True) * 100
```

```python
df_clean.head()
```

```python
print("Room types:", sorted(df_clean['reserved_room_type'].unique()))
```

```python
# Define room tier order
room_tier = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 
             'E': 5, 'F': 6, 'G': 7, 'H': 8, 'L': 9}

# Map tier numbers to each room type
df_clean['reserved_room_tier'] = df_clean['reserved_room_type'].map(room_tier)
df_clean['assigned_room_tier'] = df_clean['assigned_room_type'].map(room_tier)

# Determine direction of change
df_clean['room_change'] = df_clean.apply(
    lambda row: 'Upgraded' if row['assigned_room_tier'] > row['reserved_room_tier']
    else ('Downgraded' if row['assigned_room_tier'] < row['reserved_room_tier']
    else 'No Change'), axis=1
)

# See the results
print(df_clean['room_change'].value_counts())
print(df_clean['room_change'].value_counts(normalize=True) * 100)
```

```python
room_changes = df_clean['room_change'].value_counts()

plt.figure()
colors = ['steelblue', 'green', 'tomato']
room_changes.plot(kind='bar', color=colors)
plt.title('Room Change')
plt.xlabel('Room Change')
plt.ylabel('Amount')
plt.xticks(range(len(room_changes)), room_changes.index, rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Results
| Status | Percentage |
|---|---|
| No Change | 85.95% |
| Upgraded | 13.43% |
| Downgraded | 0.63% |

### Business Insights

**1 — Hotel honors reservations consistently**
85.62% of guests received exactly the room type they originally 
booked, indicating strong operational consistency in room 
assignment.

**2 — Upgrades significantly outweigh downgrades**
Of the 14.38% of guests who received a different room, 13.43% 
were upgraded to a better room type while only 0.63% were 
downgraded. This represents a 21:1 ratio of upgrades to 
downgrades — a positive indicator for overall guest satisfaction.

**3 — Downgrades require attention**
Despite the low percentage, 534 guests received a worse room 
than they originally reserved. These guests are at the highest 
risk of dissatisfaction and negative reviews. The hotel should 
investigate the root cause — most likely overbooking of 
specific room types.

### Limitation
This analysis assumes an alphabetical room tier order 
(A = lowest, H = highest). Without official room tier 
documentation from the hotel, upgrade and downgrade 
classifications should be interpreted with caution.

## Question 9 — Cancellation Rate

```python
# Define room tier order
room_tier = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 
             'E': 5, 'F': 6, 'G': 7, 'H': 8, 'L': 9}

# Map tier numbers to each room type
df_clean['reserved_room_tier'] = df_clean['reserved_room_type'].map(room_tier)
df_clean['assigned_room_tier'] = df_clean['assigned_room_type'].map(room_tier)

# Determine direction of change
df_clean['room_change'] = df_clean.apply(
    lambda row: 'Upgraded' if row['assigned_room_tier'] > row['reserved_room_tier']
    else ('Downgraded' if row['assigned_room_tier'] < row['reserved_room_tier']
    else 'No Change'), axis=1
)

# See the results
print(df_clean['room_change'].value_counts())
print(df_clean['room_change'].value_counts(normalize=True) * 100)
```

```python
# Sum of 1s = total cancelled
total_cancelled = df_clean['is_canceled'].sum()

# Total bookings minus cancelled = not cancelled
total_not_cancelled = len(df_clean) - total_cancelled

# Total bookings
total_bookings = len(df_clean)

# Cancellation rate
cancellation_rate = (total_cancelled / total_bookings) * 100

print(f"Total Bookings: {total_bookings:,}")
print(f"Total Cancelled: {total_cancelled:,}")
print(f"Total Not Cancelled: {total_not_cancelled:,}")
print(f"Cancellation Rate: {cancellation_rate:.2f}%")
```
